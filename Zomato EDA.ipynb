```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
## To check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
## Pie Chart - top 3 countries uses zomato

plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: Zomato Maximum Records or Transactions are from INDIA After that US then UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When Ratings are between 4.5 to 4.9  ---> Excellent
2. When Ratings are between 4.0 to 3.4 ---> Very Good
3. When Ratings are between 3.5 to 3.9 ---> good
4. When Ratings are between 3.0 to 3.4 ---> Average
5. When Ratings are between 2.5 to 2.9 ---> Average
6. When Ratings are between 2.0 to 2.4 ---> Poor

```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color', data=ratings, palette=['white','red','orange','yellow','green','green'])
```

#### Observation : 
    1. Not Rated count is very high
    2. Maximum Number of Rating are between 2.5 to 3.4

```python
### Count plot

sns.countplot(x="Rating color",data=ratings, palette=['white','red','orange','yellow','green','green'])
```
